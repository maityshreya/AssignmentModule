distan0e_moon_earth = 3.844e8 # Average distan0e between Moon and Earth in meters
Tasks
/ Cal0ulate the gravitational for0e between the Earth and the Sun

/ Compare the 0al0ulated for0es to determine whi0h gravitational for0e is stronger
/ Explain whi0h 0elestial body (Earth or Moon is more attra0ted to the other based on the 0omparison.

```python
# Constants
G = 6.67430e-11  # Gravitational constant in N * (m^2 / kg^2)

# Given masses in kilograms
mass_earth = 5.972e24
mass_moon = 7.34767309e22
mass_sun = 1.989e30

# Given distances in meters
distance_earth_sun = 1.496e11
distance_moon_earth = 3.844e8

# Calculate gravitational force between Earth and Sun
force_earth_sun = G * (mass_earth * mass_sun) / (distance_earth_sun ** 2)

# Calculate gravitational force between Moon and Earth
force_moon_earth = G * (mass_moon * mass_earth) / (distance_moon_earth ** 2)

# Print the results
print(f"Gravitational force between Earth and Sun: {force_earth_sun:.2e} N")
print(f"Gravitational force between Moon and Earth: {force_moon_earth:.2e} N")

# Compare the forces
if force_earth_sun > force_moon_earth:
    print("The gravitational force between Earth and Sun is stronger.")
else:
    print("The gravitational force between Moon and Earth is stronger.")
```

 3. Provide methods for deposit, withdraw, and balance inquiry in each subclass
 4. Test the banking system by creating instances of different account types and performing transactions.

```python
class BankAccount:
    def __init__(self, account_number, account_holder, balance=0.0):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited {amount}. New balance is {self.balance}.")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance is {self.balance}.")
        else:
            print("Insufficient funds or invalid amount.")

    def get_balance(self):
        print(f"The balance for account {self.account_number} is {self.balance}.")
        return self.balance

class SavingsAccount(BankAccount):
    def __init__(self, account_number, account_holder, balance=0.0, interest_rate=0.02):
        super().__init__(account_number, account_holder, balance)
        self.interest_rate = interest_rate

    def add_interest(self):
        interest = self.balance * self.interest_rate
        self.balance += interest
        print(f"Added interest {interest}. New balance is {self.balance}.")

class CheckingAccount(BankAccount):
    def __init__(self, account_number, account_holder, balance=0.0, overdraft_limit=100.0):
        super().__init__(account_number, account_holder, balance)
        self.overdraft_limit = overdraft_limit

    def withdraw(self, amount):
        if 0 < amount <= self.balance + self.overdraft_limit:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance is {self.balance}.")
        else:
            print("Insufficient funds or amount exceeds overdraft limit.")
# Test SavingsAccount
savings = SavingsAccount(account_number="SA123", account_holder="Alice", balance=1000.0)
savings.deposit(500)
savings.withdraw(200)
savings.get_balance()
savings.add_interest()

# Test CheckingAccount
checking = CheckingAccount(account_number="CA456", account_holder="Bob", balance=500.0, overdraft_limit=200.0)
checking.deposit(300)
checking.withdraw(700)
checking.get_balance()
checking.withdraw(200)  # This should trigger overdraft
checking.get_balance()

```

```python
class MenuItem:
    def __init__(self, name, description, price, category):
        self.__id = id(self)
        self.name = name
        self.description = description
        self.price = price
        self.category = category

    def get_id(self):
        return self.__id

    def update_info(self, name=None, description=None, price=None, category=None):
        if name:
            self.name = name
        if description:
            self.description = description
        if price:
            self.price = price
        if category:
            self.category = category

    def __str__(self):
        return f"{self.name} ({self.category}): {self.description} - Rs.{self.price:.2f}"

class FoodItem(MenuItem):
    def __init__(self, name, description, price, category, is_vegetarian=False):
        super().__init__(name, description, price, category)
        self.is_vegetarian = is_vegetarian

    def __str__(self):
        veg_info = " (Vegetarian)" if self.is_vegetarian else ""
        return super().__str__() + veg_info

class BeverageItem(MenuItem):
    def __init__(self, name, description, price, category, is_alcoholic=False):
        super().__init__(name, description, price, category)
        self.is_alcoholic = is_alcoholic

    def __str__(self):
        alc_info = " (Alcoholic)" if self.is_alcoholic else ""
        return super().__str__() + alc_info

class Menu:
    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)

    def remove_item(self, item_id):
        self.items = [item for item in self.items if item.get_id() != item_id]

    def update_item(self, item_id, name=None, description=None, price=None, category=None):
        for item in self.items:
            if item.get_id() == item_id:
                item.update_info(name, description, price, category)

    def display_menu(self):
        for item in self.items:
            print(item)

# Create menu
menu = Menu()

# Add food items
VegBiryani = FoodItem(name="Veg Biryani", description="Veg Biryani with paner", price=280,category="Main Course", is_vegetarian=True)
DumAloo = FoodItem(name="Dum Aloo", description="Dum Aloo", price=250, category="Main Course")
Daaltadka=FoodItem(name="Daal Tadka",description="Daal Tadka",price=200,category="Main Course")
menu.add_item(VegBiryani)
menu.add_item(DumAloo)
menu.add_item(Daaltadka)

# Add beverage items
coke = BeverageItem(name="Coke", description="Coca Cola", price=60, category="Soft Drink", is_alcoholic=False)
coldcoffe = BeverageItem(name="coldcoffe", description="coldcoffe", price=100, category="Healthy Beverage", is_alcoholic=False)
menu.add_item(coke)
menu.add_item(coldcoffe)

# Display menu
menu.display_menu()

# Update an item
menu.update_item(VegBiryani.get_id(), price=280)

# Remove an item
menu.remove_item(coke.get_id())

# Display updated menu
print("\nUpdated Menu:")
menu.display_menu()
```

7.Write a code for Hotel Management System using OO4S 3
& Create a Room 'lass that has attributes su'h as room number, room type, rate, and availability (private)
& Implement methods to book a room, 'he'k in a guest, and 'he'k out a guest
& Use en'apsulation to hide the room's unique identifi'ation number
& Inherit from the Room 'lass to 'reate a SuiteRoom 'lass and a StandardRoom 'lass, ea'h with their own
spe'ifi' attributes and methods.

```python
class Room:
    def __init__(self, room_number, room_type, rate):
        self.__id = id(self)
        self.room_number = room_number
        self.room_type = room_type
        self.rate = rate
        self.__availability = True

    def get_id(self):
        return self.__id

    def is_available(self):
        return self.__availability

    def book_room(self):
        if self.__availability:
            self.__availability = False
            print(f"Room {self.room_number} booked successfully.")
        else:
            print(f"Room {self.room_number} is already booked.")

    def check_in(self):
        if not self.__availability:
            print(f"Guest checked into room {self.room_number}.")
        else:
            print(f"Room {self.room_number} is not booked yet.")

    def check_out(self):
        if not self.__availability:
            self.__availability = True
            print(f"Guest checked out of room {self.room_number}. Room is now available.")
        else:
            print(f"Room {self.room_number} is already available.")

    def __str__(self):
        status = "Available" if self.__availability else "Booked"
        return f"Room {self.room_number} ({self.room_type}): Rs.{self.rate:.2f} per night - {status}"
    
class SuiteRoom(Room):
    def __init__(self, room_number, rate, has_lounge=True):
        super().__init__(room_number, "Suite", rate)
        self.has_lounge = has_lounge

    def __str__(self):
        lounge_info = "with Lounge" if self.has_lounge else "without Lounge"
        return super().__str__() + f" - {lounge_info}"

class StandardRoom(Room):
    def __init__(self, room_number, rate, has_view=False):
        super().__init__(room_number, "Standard", rate)
        self.has_view = has_view

    def __str__(self):
        view_info = "with View" if self.has_view else "without View"
        return super().__str__() + f" - {view_info}"
class Hotel:
    def __init__(self):
        self.rooms = []

    def add_room(self, room):
        self.rooms.append(room)

    def find_room_by_number(self, room_number):
        for room in self.rooms:
            if room.room_number == room_number:
                return room
        return None
    def book_room(self, room_number):
        room = self.find_room_by_number(room_number)
        if room:
            room.book_room()
        else:
            print(f"Room {room_number} not found.")

    def check_in(self, room_number):
        room = self.find_room_by_number(room_number)
        if room:
            room.check_in()
        else:
            print(f"Room {room_number} not found.")

    def check_out(self, room_number):
        room = self.find_room_by_number(room_number)
        if room:
            room.check_out()
        else:
            print(f"Room {room_number} not found.")

    def display_rooms(self):
        for room in self.rooms:
            print(room)
# Create hotel
hotel = Hotel()

# Add rooms
suite1 = SuiteRoom(room_number=101, rate=2500.0, has_lounge=True)
standard1 = StandardRoom(room_number=102, rate=1800.0, has_view=True)
standard2 = StandardRoom(room_number=103, rate=1300.0, has_view=False)
hotel.add_room(suite1)
hotel.add_room(standard1)
hotel.add_room(standard2)

# Display rooms
hotel.display_rooms()

# Book a room
hotel.book_room(101)

# Check in a guest
hotel.check_in(101)

# Check out a guest
hotel.check_out(101)

# Try to book a room that's already booked
hotel.book_room(101)
# Try to check in a guest to an available room
hotel.check_in(102)

# Display updated room status
print("\nUpdated Room Status:")
hotel.display_rooms()
```

```python
class Member:
    def __init__(self, name, age, membership_type):
        self.__id = id(self)
        self.name = name
        self.age = age
        self.membership_type = membership_type
        self.__membership_status = "Active"

    def get_id(self):
        return self.__id

    def get_membership_status(self):
        return self.__membership_status

    def register(self):
        print(f"Member {self.name} registered successfully with {self.membership_type} membership.")

    def renew_membership(self):
        if self.__membership_status == "Active":
            print(f"Membership for {self.name} is already active.")
        else:
            self.__membership_status = "Active"
            print(f"Membership for {self.name} renewed successfully.")

    def cancel_membership(self):
        if self.__membership_status == "Cancelled":
            print(f"Membership for {self.name} is already cancelled.")
        else:
            self.__membership_status = "Cancelled"
            print(f"Membership for {self.name} cancelled successfully.")

    def __str__(self):
        return f"Member: {self.name}, Age: {self.age}, Membership Type: {self.membership_type}, Status: {self.__membership_status}"

class FamilyMember(Member):
    def __init__(self, name, age, membership_type, family_discount):
        super().__init__(name, age, membership_type)
        self.family_discount = family_discount

    def __str__(self):
        return super().__str__() + f", Family Discount: {self.family_discount}%"

class IndividualMember(Member):
    def __init__(self, name, age, membership_type, personal_trainer=False):
        super().__init__(name, age, membership_type)
        self.personal_trainer = personal_trainer

    def __str__(self):
        trainer_info = "with Personal Trainer" if self.personal_trainer else "without Personal Trainer"
        return super().__str__() + f", {trainer_info}"

class FitnessClub:
    def __init__(self):
        self.members = []

    def add_member(self, member):
        self.members.append(member)
        member.register()

    def find_member_by_id(self, member_id):
        for member in self.members:
            if member.get_id() == member_id:
                return member
        return None

    def renew_membership(self, member_id):
        member = self.find_member_by_id(member_id)
        if member:
            member.renew_membership()
        else:
            print(f"Member with ID {member_id} not found.")

    def cancel_membership(self, member_id):
        member = self.find_member_by_id(member_id)
        if member:
            member.cancel_membership()
        else:
            print(f"Member with ID {member_id} not found.")

    def display_members(self):
        for member in self.members:
            print(member)

# Create fitness club
club = FitnessClub()

# Add members
family_member = FamilyMember(name="JayKumar", age=40, membership_type="Family", family_discount=15)
individual_member = IndividualMember(name="Rajiv", age=30, membership_type="Individual", personal_trainer=True)
club.add_member(family_member)
club.add_member(individual_member)

# Display members
club.display_members()

# Renew a membership
club.renew_membership(family_member.get_id())

# Cancel a membership
club.cancel_membership(individual_member.get_id())

# Try to renew a cancelled membership
club.renew_membership(individual_member.get_id())

# Display updated member status
print("\nUpdated Member Status:")
club.display_members()
```

```python
class Event:
    def __init__(self, name, date, time, location):
        self.__id = id(self)
        self.name = name
        self.date = date
        self.time = time
        self.location = location
        self.__attendees = []

    def get_id(self):
        return self.__id

    def add_attendee(self, attendee):
        if attendee not in self.__attendees:
            self.__attendees.append(attendee)
            print(f"Attendee {attendee} added successfully.")
        else:
            print(f"Attendee {attendee} is already on the list.")

    def remove_attendee(self, attendee):
        if attendee in self.__attendees:
            self.__attendees.remove(attendee)
            print(f"Attendee {attendee} removed successfully.")
        else:
            print(f"Attendee {attendee} is not on the list.")

    def get_total_attendees(self):
        return len(self.__attendees)

    def __str__(self):
        return f"Event: {self.name}, Date: {self.date}, Time: {self.time}, Location: {self.location}, Total Attendees: {self.get_total_attendees()}"
    
class PrivateEvent(Event):
    def __init__(self, name, date, time, location, host):
        super().__init__(name, date, time, location)
        self.host = host

    def __str__(self):
        return super().__str__() + f", Host: {self.host}"

class PublicEvent(Event):
    def __init__(self, name, date, time, location, organizer):
        super().__init__(name, date, time, location)
        self.organizer = organizer

    def __str__(self):
        return super().__str__() + f", Organizer: {self.organizer}"


class EventManagement:
    def __init__(self):
        self.events = []

    def create_event(self, event):
        self.events.append(event)
        print(f"Event '{event.name}' created successfully.")

    def find_event_by_id(self, event_id):
        for event in self.events:
            if event.get_id() == event_id:
                return event
        return None

    def add_attendee(self, event_id, attendee):
        event = self.find_event_by_id(event_id)
        if event:
            event.add_attendee(attendee)
        else:
            print(f"Event with ID {event_id} not found.")

    def remove_attendee(self, event_id, attendee):
        event = self.find_event_by_id(event_id)
        if event:
            event.remove_attendee(attendee)
        else:
            print(f"Event with ID {event_id} not found.")

    def display_events(self):
        for event in self.events:
            print(event)


# Create event management system
management = EventManagement()

# Create events
private_event = PrivateEvent(name="Birthday Party", date="2024-08-10", time="18:00", location="Raj's House", host="Raj Verma")
public_event = PublicEvent(name="Tech Conference", date="2024-09-15", time="09:00", location="Convention Center", organizer="Tech Org")

# Add events to management system
management.create_event(private_event)
management.create_event(public_event)

# Add attendees
management.add_attendee(private_event.get_id(), "Ananya")
management.add_attendee(private_event.get_id(), "Bhavik")
management.add_attendee(public_event.get_id(), "Chetan")

# Remove an attendee
management.remove_attendee(private_event.get_id(), "Ananya")

# Display events
management.display_events()
```

10.Write a code for Airline Reservation System using OO4S3
& Create a Flight 'lass that has attributes su'h as flight number, departure and arrival airports, departure and
arrival times, and available seats (private)
& Implement methods to book a seat, 'an'el a reservation, and get the remaining available seats
& Use en'apsulation to hide the flight's unique identifi'ation number
& Inherit from the Flight 'lass to 'reate a Domesti'Flight 'lass and an InternationalFlight 'lass, ea'h with their
own spe'ifi' attributes and methods.

```python
class Flight:
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats):
        self.__id = id(self)
        self.flight_number = flight_number
        self.departure_airport = departure_airport
        self.arrival_airport = arrival_airport
        self.departure_time = departure_time
        self.arrival_time = arrival_time
        self.__available_seats = total_seats
        self.__total_seats = total_seats

    def get_id(self):
        return self.__id

    def book_seat(self):
        if self.__available_seats > 0:
            self.__available_seats -= 1
            print(f"Seat booked successfully. Remaining seats: {self.__available_seats}")
        else:
            print("No available seats left.")

    def cancel_reservation(self):
        if self.__available_seats < self.__total_seats:
            self.__available_seats += 1
            print(f"Reservation cancelled successfully. Remaining seats: {self.__available_seats}")
        else:
            print("No reservations to cancel.")

    def get_remaining_seats(self):
        return self.__available_seats

    def __str__(self):
        return f"Flight Number: {self.flight_number}, Departure: {self.departure_airport} at {self.departure_time}, Arrival: {self.arrival_airport} at {self.arrival_time}, Remaining Seats: {self.__available_seats}"

class DomesticFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats, in_flight_snack):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats)
        self.in_flight_snack = in_flight_snack

    def __str__(self):
        return super().__str__() + f", In-flight Snack: {self.in_flight_snack}"

class InternationalFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats, visa_required):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats)
        self.visa_required = visa_required

    def __str__(self):
        return super().__str__() + f", Visa Required: {self.visa_required}"

class AirlineReservationSystem:
    def __init__(self):
        self.flights = []

    def add_flight(self, flight):
        self.flights.append(flight)
        print(f"Flight '{flight.flight_number}' added successfully.")

    def find_flight_by_id(self, flight_id):
        for flight in self.flights:
            if flight.get_id() == flight_id:
                return flight
        return None

    def book_seat(self, flight_id):
        flight = self.find_flight_by_id(flight_id)
        if flight:
            flight.book_seat()
        else:
            print(f"Flight with ID {flight_id} not found.")

    def cancel_reservation(self, flight_id):
        flight = self.find_flight_by_id(flight_id)
        if flight:
            flight.cancel_reservation()
        else:
            print(f"Flight with ID {flight_id} not found.")

    def display_flights(self):
        for flight in self.flights:
            print(flight)
            
# Create airline reservation system
system = AirlineReservationSystem()

# Create flights
domestic_flight = DomesticFlight(flight_number="DF123", departure_airport="JFK", arrival_airport="LAX", departure_time="2024-08-10 08:00", arrival_time="2024-08-10 11:00", total_seats=100, in_flight_snack=True)
international_flight = InternationalFlight(flight_number="IF456", departure_airport="JFK", arrival_airport="LHR", departure_time="2024-09-15 20:00", arrival_time="2024-09-16 08:00", total_seats=200, visa_required=True)

# Add flights to system
system.add_flight(domestic_flight)
system.add_flight(international_flight)

# Book seats
system.book_seat(domestic_flight.get_id())
system.book_seat(domestic_flight.get_id())
system.book_seat(international_flight.get_id())

# Cancel a reservation
system.cancel_reservation(domestic_flight.get_id())

# Display flights
system.display_flights()

```
